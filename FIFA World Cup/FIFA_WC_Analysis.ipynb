```python
print("Match`Outcomes by home and away teams\n\n")
def label(matches):
    if matches["Home Team Goals"] > matches["Away Team Goals"]:
        return "Home team win"
    if matches["Away Team Goals"] > matches["Home Team Goals"]:
        return "Away team win"
    if matches["Home Team Goals"] == matches["Away Team Goals"]:
        return "DRAW"

matches["outcome"] = matches.apply(lambda matches:label(matches),axis=1)
hw=matches[matches['outcome']=="Home team win"]['outcome'].count()/(matches['outcome'].count())
aw=matches[matches['outcome']=="Away team win"]['outcome'].count()/(matches['outcome'].count())
dr=matches[matches['outcome']=="DRAW"]['outcome'].count()/(matches['outcome'].count())
print("Home win percentage is ",hw*100)
print("Away win percentage is ",aw*100)
print("Draw percentage is ",dr*100)
```

```python
def win(matches):
    if matches["Home Team Goals"] > matches["Away Team Goals"]:
        return matches["Home Team Name"]
    if matches["Away Team Goals"] > matches["Home Team Goals"]:
        return matches["Away Team Name"]
    if matches["Home Team Goals"] == matches["Away Team Goals"]:
        return None

def loss(matches):
    if matches["Home Team Goals"] < matches["Away Team Goals"]:
        return matches["Home Team Name"]
    if matches["Away Team Goals"] < matches["Home Team Goals"]:
        return matches["Away Team Name"]
    if matches["Home Team Goals"] == matches["Away Team Goals"]:
        return None
matches["Winning Team"] = matches.apply(lambda matches:win(matches),axis=1)
matches["Losing Team"]=matches.apply(lambda matches:loss(matches),axis=1)
matches
```

```python
print("Penalty Analysis")
penalty=pd.read_csv('penalties.csv',encoding='iso-8859-1')
def clean(data):
    data["F"] = data["F"].str.replace("","-")
    data["S"] = data["S"].str.replace("","-")
    data["Winner"] = data["Winner"].str.replace("West Germany","Germany")
    data["winner_penalties"] = data["S"].str.split("-").str[0]
    data["loser_penalties"]  = data["S"].str.split("-").str[1]
    data["Date"]   = pd.to_datetime(data["Date"],format="%d-%b-%y") 
    data["year"]   = pd.DatetimeIndex(data["Date"]).year
    data = data.rename(columns = {"S":"penalties_scored","F":"final_score"})
    return data

penalty=clean(penalty)
penalty=penalty.reindex(columns=['year','Date','Round','Winner','Loser','final_score','winner_penalties','loser_penalties'])
penalty
```

```python
print("penalty matches by year")
py=penalty['year'].value_counts().reset_index()
py.columns=['year','shootouts']
py.set_index('year',inplace=True)
py.sort_index(inplace=True)
py
```

```python
print("Penalty matches by round")
pr=penalty['Round'].value_counts().reset_index()
pr.columns=['Round','shootouts']
pr.set_index('Round',inplace=True)
pr.sort_index(inplace=True)
pr
```

```python
print("Penalty shootouts in final")
p_final=penalty[penalty['Round']=='Final']
p_final
```

```python
print("Most penalties played by countries")
pc=pd.concat([penalty["Winner"],penalty["Loser"]])
pc=pc.value_counts().reset_index()
pc.columns=['Country','Penalty Shootouts']
pc=pc.set_index('Country')
pc
```

```python
print("Countries with most penalty shootout win")
pw=penalty['Winner'].value_counts().reset_index()
pw.columns=['Country','Matches won']
pw.set_index('Country',inplace=True)
pw
```

```python
print("Countries with most penalty shootout loses")
pl=penalty['Loser'].value_counts().reset_index()
pl.columns=['Country','Matches lost']
pl.set_index('Country',inplace=True)
pl
```

```python
print("Penalty shootout result of each country")
presult=pd.merge(pw,pl,left_index=True,right_index=True,how="outer")
presult=presult.fillna(0)
presult
```
